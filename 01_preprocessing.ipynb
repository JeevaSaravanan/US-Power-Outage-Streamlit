```python
import pandas as pd
```

```python
data_14 = pd.read_csv("data/eaglei_outages_with_events_2014.csv")
data_15 = pd.read_csv("data/eaglei_outages_with_events_2015.csv")
data_16 = pd.read_csv("data/eaglei_outages_with_events_2016.csv")
data_17 = pd.read_csv("data/eaglei_outages_with_events_2017.csv")
data_18 = pd.read_csv("data/eaglei_outages_with_events_2018.csv")
data_19 = pd.read_csv("data/eaglei_outages_with_events_2019.csv")
data_20 = pd.read_csv("data/eaglei_outages_with_events_2020.csv")
data_21 = pd.read_csv("data/eaglei_outages_with_events_2021.csv")
data_22 = pd.read_csv("data/eaglei_outages_with_events_2022.csv")
data_23 = pd.read_csv("data/eaglei_outages_with_events_2023.csv")
```

```python
data = pd.concat([data_14,data_15,data_16,data_17,data_18,data_19,data_20,data_21,data_22,data_23],axis=0,ignore_index=True)
data
```

```python
data.drop(['event_id','state_event','Datetime Event Began','Datetime Restoration'],inplace=True,axis=1)
```
